Build a regression model.

```python
# Import necessary libraries
import pandas as pd  # For data handling
import statsmodels.api as sm  # For regression modeling
import matplotlib.pyplot as plt  # For visualization

# Step 1: Load the dataset
# Load the merged Yelp dataset
merged_yelp = pd.read_csv(r"C:\Users\Admin\Downloads\statistical modelling project\-Statistical-Modelling-Project\merged_yelp_data.csv")
print("Merged Yelp DataFrame:")
print(merged_yelp.head())  # Display the first few rows of the dataset

# Step 2: Data Preprocessing
# Select relevant columns for the regression model
model_data = merged_yelp[["available_bikes", "rating", "review_count"]].dropna()  # Drop rows with missing values

# Display the preprocessed data
print("\nPreprocessed Data for the Regression Model:")
print(model_data.head())

# Step 3: Define Variables
# Dependent variable (y) - Number of available bikes
y = model_data["available_bikes"]

# Independent variables (X) - POI rating and review count
X = model_data[["rating", "review_count"]]

# Add a constant term to the independent variables for the regression model
X = sm.add_constant(X)

# Step 4: Build the Regression Model
# Fit an Ordinary Least Squares (OLS) regression model
model = sm.OLS(y, X).fit()

# Display the model summary
print("\nRegression Model Summary:")
print(model.summary())

# Step 5: Visualize Actual vs. Predicted Values
# Predict the values using the regression model
y_pred = model.predict(X)

# Scatter plot of actual vs. predicted values
plt.figure(figsize=(8, 5))
plt.scatter(y, y_pred, alpha=0.7, color="green")
plt.title("Actual vs. Predicted Values")
plt.xlabel("Actual Number of Available Bikes")
plt.ylabel("Predicted Number of Available Bikes")
plt.grid(True)
plt.show()
```
